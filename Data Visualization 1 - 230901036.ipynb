```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([0,10])
ypoints=np.array([20,100])
print(plt.plot(xpoints,ypoints))
print(plt.show())
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([0,5])
y=np.array([0,25])
z=plt.plot(x,y,marker='p')
print(z)
print(plt.show())
```

```python
xpoints=([1,2,4,6,7])
ypoints=([0,2,8,6,6])
z=plt.plot(xpoints,ypoints,marker='p',color='black')
print(z)
print(plt.plot(xpoints,ypoints))
print(plt.show())
```
